【問題5】リストへの保存
ここまでは43回折った後の最後の値だけを使用していましたが、グラフで可視化するためには過程の値も必要です。for文を使用したコードに、過程の値合計44個を記録するコードを加えてください。

```python
THICKNESS = 0.00008
folded_thickness = THICKNESS * 2
folded_thickness_list = []
folded_thickness_list.append(THICKNESS)

for i in range(1,44):
    THICKNESS = THICKNESS * 2
    folded_thickness_list.append(THICKNESS)

print(folded_thickness_list)
print(len(folded_thickness_list))
```

【問題6】折れ線グラフの表示

```python
THICKNESS = 0.00008
folded_thickness = THICKNESS * 2
folded_thickness_list = []
folded_thickness_list.append(THICKNESS)

for i in range(1,44):
    THICKNESS = THICKNESS * 2
    folded_thickness_list.append(THICKNESS)
    
import matplotlib.pyplot as plt
%matplotlib inline
# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness_list)  # 「リスト名」のところにリストの変数名を入れる
plt.show()  # com + con + enter
```

【問題7】グラフのカスタマイズ

```python
#グラフの線を黄色にする
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness_list, color='yellow')
plt.show()

#軸の値を太くする
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.tick_params(labelsize=15) # 軸の値に関する設定を行う
plt.plot(folded_thickness_list)
plt.show()

#メモリを内側と外側に表示
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.tick_params(direction = 'inout') # 軸の値に関する設定を行う
plt.plot(folded_thickness_list)
plt.show()

#赤の点線グラフ
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.tick_params(direction = 'inout') # 軸の値に関する設定を行う
plt.plot(folded_thickness_list,'--r')
plt.show()
```
